```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
np.random.seed(42)
n_sample = 50
```

```python
cats_x1 = np.random.uniform(1,5,n_sample)
cats_x2 = np.random.uniform(1,5,n_sample)
cats = np.column_stack((cats_x1,cats_x2))
cats_label = -np.ones(n_sample)
```

```python
dogs_x1 = np.random.uniform(4,8,n_sample)
dogs_x2 = np.random.uniform(4,8,n_sample)
dogs = np.column_stack((dogs_x1,dogs_x2))
dogs_label = np.ones(n_sample)
```

```python
x= np.vstack((cats,dogs))
y= np.concatenate((cats_label,dogs_label))
```

```python
plt.figure(figsize= (20,9),dpi= 100)
plt.subplot (1,2,1)
plt.scatter(cats[:,0],cats[:,1],color= 'red',label= 'cats')
plt.scatter(dogs[:,0],dogs[:,1],color= 'blue',label= 'dogs')
plt.legend()
```

```python
x_bais = np.column_stack([np.ones(len(x)),x])
```

```python
np.random.seed(2)
w = np.random.rand(3)

eta = 0.001
epochs = 100000
costval =np.zeros(epochs)
```

```python
for i in range (epochs):
  z = x_bais @ w
  errors = z- y
  w =w - eta * (x_bais.T @ errors)/len(x)
  costval[i]= 0.5* np.mean(errors**2)
  if costval[i] < 1e-5 :
    break
```

```python
b = w[0]
w1 = w[1]
w2 =w[2]
```

```python
x1_val = np.linspace(0,9,100)
x2_val = -(b+w1*x1_val)/w2

```

```python
plt.plot(x1_val,x2_val,'-r',label = "decision Boundry")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.title("Decision Boundry")
plt.grid(True ,alpha = 0.3)
plt.xlim(0,9)
plt.ylim(0,9)

```
